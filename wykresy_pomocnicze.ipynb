```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_imu = 50.0
f_inf = 5.0
N = 25
t_end = 2.0

imu_dt = 1.0 / f_imu
inf_dt = 1.0 / f_inf
win_len = N * imu_dt

t_imu = np.arange(0.0, t_end + 1e-12, imu_dt)
t_inf = np.arange(0.0, t_end + 1e-12, inf_dt)

k = min(5, len(t_inf) - 1)
t_k = t_inf[k]
t0 = max(0.0, t_k - win_len)
t1 = t_k

fig, ax = plt.subplots(figsize=(11, 3.8))

y_imu = 1.0
y_inf = 0.0

ax.hlines(y_imu, 0, t_end, linewidth=1)
ax.hlines(y_inf, 0, t_end, linewidth=1)

ax.vlines(t_imu, y_imu - 0.12, y_imu + 0.12, linewidth=1)
ax.vlines(t_inf, y_inf - 0.18, y_inf + 0.18, linewidth=2)

ax.axvspan(t0, t1, alpha=0.2)
ax.vlines([t0, t1], -0.35, 1.25, linestyles="--", linewidth=1)

ax.text(-0.02, y_imu, "IMU (50 Hz)", ha="right", va="center")
ax.text(-0.02, y_inf, "Pętla wnioskowania (5 Hz)", ha="right", va="center")

# estetyka osi
ax.set_xlim(0, t_end)
ax.set_ylim(-0.6, 1.6)
ax.set_yticks([])
ax.set_xlabel("Czas [s]")
ax.grid(True, axis="x", alpha=0.3)

ax.set_title("Diagram czasowy: IMU (50 Hz) vs pętla wnioskowania (5 Hz) oraz okno N=25 próbek")

plt.tight_layout()
plt.savefig("rys_5_2_diagram_czasowy.pdf")
plt.savefig("rys_5_2_diagram_czasowy.png", dpi=300)
plt.show()
```
